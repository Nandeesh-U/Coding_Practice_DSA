Find the GCD of two given positive integers using recursion.

We will use euclidean algorithm to find the gcd.

gcd(m,n) = gcd((m%n),m)

```python
def gcd(m,n):
  assert (m>0 and n>0) and (int(m)==m and int(n)==n), 'Both the numbers should be positive integers only'
  m,n = sorted([m,n])
  if n%m == 0:
    return m
  else:
    return gcd((n%m),m)
```
